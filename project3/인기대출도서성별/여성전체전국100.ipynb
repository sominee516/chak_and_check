```python
# 전국통계는 30위까지의 결과가 어린이분야 도서가 대부분이어서 100위까지 범위를 확대해 성별로 비교해보았다

import pandas as pd
e_s=pd.read_excel('BestLoanList_여성_전체_전국_100.xlsx')
e_s = e_s.head(100)
import collections
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
plt.rcParams['font.family'] = 'Malgun Gothic'
# 나열 순서는 다르지만 분야별 통일감을 주기 위해 분야별 같은 색상을 설정해줌
palette = ['#ea96a3','#83a846','#50acbe','#d97157','#4cabac','#ce974c','#e6957a','#9da045','#8f88bd'] ## 색상 지정
plt.xticks(fontsize=12,rotation=25)
plt.yticks(fontsize=12)
plt.ylim(0, 100)
ax = sns.countplot(x='분야',data=e_s,palette=palette)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['left'].set_visible(False)
plt.gca().set_facecolor('#f0efed')
ax.set(xlabel=None)
plt.savefig('busanfemale100.png')
plt.show()
```
