```python
import pandas as pd
e_s=pd.read_excel('인기대출도서2019.xlsx')
e_s = e_s.head(100)
import collections
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
# 인기대출목록 그래프에서는 같은 분야 같은 색상 설정
palette = ['#ea96a3','#83a846','#ce974c','#e6957a','#50acbe','#845b53','#4cabac','#edda80','#8f88bd','#9da045'] ## 색상 지정
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.figure(figsize=(20,10))
plt.title('2019 인기 대출 도서',fontsize=40)
plt.xticks(fontsize=20,rotation=30)
plt.yticks(fontsize=20)
# 쉬운 비교를 위해 y축 한계치를 정해줌
plt.ylim(0, 40)
ax = sns.countplot(x='분야',data=e_s,palette=palette)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['left'].set_visible(False)
plt.gca().set_facecolor('#f0efed')
ax.set(xlabel=None)
ax.set(ylabel=None)
plt.savefig('2019bestloan.png')
plt.show()
```
