```python
import pandas as pd
e_s=pd.read_excel('인기대출도서2020.xlsx')
e_s = e_s.head(100)
import collections
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
palette = ['#83a846','#ea96a3','#48b058','#ce974c','#e6957a','#50acbe','#4cabac','#8f88bd','#845b53','#9da045'] ## 색상 지정
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.figure(figsize=(20,10))
plt.title('2020 인기 대출 도서',fontsize=40)
# plt.rc('axes', labelsize=30)   # x,y축 label 폰트 크기
# plt.rc('xtick', labelsize=20)
# plt.xticks(rotation=45)
plt.xticks(fontsize=20,rotation=30)
plt.yticks(fontsize=20)
plt.ylim(0, 40)
ax = sns.countplot(x='분야',data=e_s,palette=palette)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['left'].set_visible(False)
plt.gca().set_facecolor('#f0efed')
ax.set(xlabel=None)
ax.set(ylabel=None)
plt.savefig('2020bestloan.png')
plt.show()
```
