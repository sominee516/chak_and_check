```python
import pandas as pd
e_s=pd.read_excel('교보 2021년 상반기 베스트셀러 목록 100위.xlsx')
e_s = e_s.head(100)
```

```python
# 50대의 도서구매 비율
age = e_s['연령 1위'] == '50대'
# 여자
female = e_s['성별 여']
female50 = e_s[age & female]
female50
```

```python
# 컬럼 수정
e_s.columns=['순위','ISBN','도서명','여','남','1위','2위','저자','출판사','분야']
e_s
```

```python
# 여,남 int형으로 형변환하기위해 %잘라내기
e_s.여 = e_s.여.str.split('%').str.get(0)
e_s.남 = e_s.남.str.split('%').str.get(0)
e_s
```

```python
# 결측치 0으로 바꾸기
e_s['여'] = e_s['여'].fillna(0)
e_s['1위'] = e_s['1위'].fillna(0)
e_s['남'] = e_s['남'].fillna(0)
e_s['2위'] = e_s['2위'].fillna(0)
e_s
```

```python
# 그래프로 카운트하기 위해 형변환 
e_s.남 = e_s.남.astype('int')
e_s.여 = e_s.여.astype('int')
e_s.info()
```

```python
# 50대 여성 중에서도 여성점유율이 50%이상인 값을 가져오기
age50 = e_s['1위'] == '50대'
female = e_s['여'] >= 50
```

```python
female50 = e_s[age50 & female]
female50
```

```python
female50["성별"] = "여"
female50
```

```python
age50 = e_s['1위'] == '50대'
male = e_s['남'] >= 50
male50 = e_s[age50 & male]
male50
```

```python
male50["성별"] = "남"
male50
```

```python
# 여자, 남자 뽑아온 데이터프레임을 비교하기 위해 다시 합치기
result2 = pd.concat([female50,male50], ignore_index=True)
result2
```

```python
import collections
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['font.family'] = 'Malgun Gothic'
palette = ['lightpink','lightblue']
ax = sns.countplot(x='분야', hue='성별',data=result2,palette=palette)
# plt.figure(figsize=(20,10))
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['left'].set_visible(False)
plt.gca().set_facecolor('#f0efed')
ax.set(xlabel=None)
# plt.savefig('bestloan50.png')
plt.title('50대 성별 선호 분야')
plt.savefig('gender50.png')
plt.show()
```
