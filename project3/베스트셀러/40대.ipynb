```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Malgun Gothic'
import warnings
warnings.filterwarnings('ignore')
```

```python

e_s=pd.read_excel('교보 2021년 상반기 베스트셀러 목록 100위.xlsx')
e_s = e_s.head(100)
```

```python
#is_asia = df['continent'] == 'Asia'
age40 = e_s['연령 1위'] == '40대'
female = e_s['성별 여']

female40 = e_s[age40 & female]
female40
```

```python
e_s.info()
```

```python
# data.columns=['행정구역', '총인구수', '세대수', '세대당인구', '남자','여자', '남여비율']
e_s.columns=['순위','ISBN','도서명','여','남','1위','2위','저자','출판사','분야']
e_s
```

```python
# data.행정구역 = data.행정구역.str.split('(').str.get(0)
e_s.여 = e_s.여.str.split('%').str.get(0)
e_s.남 = e_s.남.str.split('%').str.get(0)
e_s
```

```python
e_s.info()
```

```python
# For a single column using Pandas: df['DataFrame Column'] = df['DataFrame Column'].fillna(0)
# df.fillna(0)
e_s['여'] = e_s['여'].fillna(0)
e_s['1위'] = e_s['1위'].fillna(0)
e_s['남'] = e_s['남'].fillna(0)
e_s['2위'] = e_s['2위'].fillna(0)
e_s
```

```python
e_s.남 = e_s.남.astype('int')
e_s.여 = e_s.여.astype('int')
e_s.info()
```

```python
### 30,40대는 아동분야 책이 많았음 -> 부모의 자녀도서구매를 확인할 수 있었음
# 그래서 자녀책 구매를 제외한 30,40대의 구매를 알아보기 위해 아동 관련 분야를 배제함
age40 = e_s['1위'] == '40대'
female = e_s['여'] >= 70
kids = e_s['분야'] != '아동'
kidsc = e_s['분야'] != '아동만화'
teen = e_s['분야'] != '청소년'
```

```python
#베스트 셀러 _40대 여성이 제일 많이 보는 책 _ 도서 8개 아동도서가 7
female40 = e_s[age40 & female & kids & kidsc & teen]
#female40 = female40.reset_index(drop =False, inplace = True)
female40
```

```python
#베스트 셀러 _40대 여성이 제일 많이 보는 책 _ 도서 19권 아동도서가 8권 청소년 1권
female40 = e_s[age40 & female ]
#female40 = female40.reset_index(drop =False, inplace = True)
female40.count()
female40count = female40.shape[0]
```
