```python
import pandas as pd
e_s=pd.read_excel('교보 2021년 상반기 베스트셀러 목록 100위.xlsx')
e_s = e_s.head(100)
```

```python
#is_asia = df['continent'] == 'Asia'
age = e_s['연령 1위'] == '50대'
female = e_s['성별 여']

female50 = e_s[age & female]
female50
```

```python
e_s.columns=['순위','ISBN','도서명','여','남','1위','2위','저자','출판사','분야']
e_s
```

```python
e_s.여 = e_s.여.str.split('%').str.get(0)
e_s.남 = e_s.남.str.split('%').str.get(0)
e_s
```

```python
e_s['여'] = e_s['여'].fillna(0)
e_s['1위'] = e_s['1위'].fillna(0)
e_s['남'] = e_s['남'].fillna(0)
e_s['2위'] = e_s['2위'].fillna(0)
e_s
```

```python
e_s.남 = e_s.남.astype('int')
e_s.여 = e_s.여.astype('int')
e_s.info()
```

```python
age50 = e_s['1위'] == '50대'
female = e_s['여'] >= 50
```

```python
female50 = e_s[age50 & female]
#female40 = female40.reset_index(drop =False, inplace = True)
female50
```

```python
female50["성별"] = "여"
female50
```

```python
age50 = e_s['1위'] == '50대'
male = e_s['남'] >= 50
male50 = e_s[age50 & male]
#female40 = female40.reset_index(drop =False, inplace = True)
male50
```

```python
male50["성별"] = "남"
male50
```

```python
result2 = pd.concat([female50,male50], ignore_index=True)
result2
```

```python
import collections
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['font.family'] = 'Malgun Gothic'
sns.countplot(x='분야', hue='성별',data=result2)
```
