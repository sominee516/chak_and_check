```python
import pandas as pd
e_s=pd.read_excel('교보 2021년 상반기 베스트셀러 목록 100위.xlsx')
e_s = e_s.head(100)
```

```python
#is_asia = df['continent'] == 'Asia'
age = e_s['연령 1위'] == '20대'
female = e_s['성별 여']

female20 = e_s[age & female]
female20
```

```python
e_s.columns=['순위','ISBN','도서명','여','남','1위','2위','저자','출판사','분야']
e_s
```

```python
e_s.여 = e_s.여.str.split('%').str.get(0)
e_s.남 = e_s.남.str.split('%').str.get(0)
e_s
```

```python
e_s['여'] = e_s['여'].fillna(0)
e_s['1위'] = e_s['1위'].fillna(0)
e_s['남'] = e_s['남'].fillna(0)
e_s['2위'] = e_s['2위'].fillna(0)
e_s
```

```python
e_s.남 = e_s.남.astype('int')
e_s.여 = e_s.여.astype('int')
e_s.info()
```

```python
age20 = e_s['1위'] == '20대'
female = e_s['여'] >= 50
kids = e_s['분야'] != '아동'
kidsc = e_s['분야'] != '아동만화'
teen = e_s['분야'] != '청소년'
```

```python

female20 = e_s[age20 & female & kids & kidsc & teen]
#female40 = female40.reset_index(drop =False, inplace = True)
female20
```

```python
#20대 여자 : 언어, 말투, 심리 관심사

female20 = e_s[age20 & female ]
#female40 = female40.reset_index(drop =False, inplace = True)
female20.count()
female20
```
